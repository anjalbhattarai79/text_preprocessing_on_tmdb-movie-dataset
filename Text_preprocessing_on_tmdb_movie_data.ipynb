```python
import requests
import time
import json

API_KEY = "YOUR_API_KEY"
BASE_URL = "https://api.themoviedb.org/3/movie/top_rated"
LANGUAGE = "en-US"

all_movies = []
total_pages = 400  # or dynamically fetch from API

for page in range(1, total_pages + 1):
    url = f"{BASE_URL}?api_key={API_KEY}&language={LANGUAGE}&page={page}"
    response = requests.get(url)

    # Check request success
    if response.status_code == 200:
        data = response.json()
        all_movies.extend(data.get("results", []))
        print(f"✅ Fetched page {page} / {total_pages}")
    else:
        print(f"⚠️ Failed at page {page}: {response.status_code}")
        break

    # Avoid hitting API rate limit
    time.sleep(0.025)

# Save all movie details to JSON file
with open("top_rated_movies.json", "w", encoding="utf-8") as f:
    json.dump(all_movies, f, ensure_ascii=False, indent=4)

print(f"\n✅ Total movies fetched: {len(all_movies)}")
```

# Text-Preprocessing on overview column


### 1. LowerCasing

```python
movie_data['overview'] = movie_data['overview'].str.lower()
```

### 2. Spelling Correction

```python
from textblob import TextBlob
def spelling_correction(text):
  textBlb = TextBlob(text)
  return textBlb.correct().string
```

```python
# INTERRUPTED AS TIME HEAVY PROCESS

movie_data['overview'] = movie_data['overview'].apply(spelling_correction)
```

#### 3. Punctuation {Symbols} Removal



```python
import string
string.punctuation

exclude = string.punctuation

def remove_punc(text):
    for char in exclude:
        text = text.replace(char,'')
    return text
```

```python
movie_data['overview'] = movie_data['overview'].apply(remove_punc)
```

#### 4. Stop Word Removal

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
movie_data['overview'] = movie_data['overview'].apply(remove_stopwords)
movie_data
```

#### 5. Tokenization with ntlk.tokenize

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
movie_data['tokens_by_nltk'] = movie_data['overview'].apply(lambda x: word_tokenize(x))
movie_data
```

#### 6. Tokenization with spacy

SPACY supports:
- Tokenization
- Lemmatization
- Part-of-speech (POS) tagging
- Dependency parsing
- Named Entity Recognition (NER)

```python
import spacy
nlp = spacy.load('en_core_web_sm')

movie_data['tokens_by_spacy'] = movie_data['overview'].apply(lambda text: [token.text for token in nlp(text)])
movie_data
```

#### 7. Stemming <br>
Note : The output root word of stemming itslef don't neccessarily carry actual meaning. <br>
--  To solve this *lemmatization* can also be implemented.

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return y

movie_data['tokens_by_nltk'] = movie_data['overview'].apply(stem)
movie_data['tokens_by_spacy'] = movie_data['overview'].apply(stem)
```
